```python
import numpy as np
import pandas as pd
import string
pd.options.mode.chained_assignment = None

full_df = pd.read_csv("D:/kuliah/Skripsi/ipynb/Data/textpreprocessing2.csv", nrows=56000)
df = full_df[["Text"]]
df["Text"] = df["Text"].astype(str)

full_df.head()

#lowercase / case folding
df["text_lower"] = df["Text"].str.lower()
df.head()

# Tokenizing
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    # """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["text_wo_punct"] = df["text_lower"].apply(lambda text: remove_punctuation(text))
df.head()

from collections import Counter
cnt = Counter()
for text in df["text_lower"].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(10)

FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

# Filtering
df["text_wo_stopfreq"] = df["text_wo_punct"].apply(lambda text: remove_freqwords(text))
df.head()

from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
# stemming
df["text_stemmed"] = df["text_wo_stopfreq"].apply(lambda text: stem_words(text))
df.head()

# df.to_csv('textpreprocessing3.csv')
```
